```python
import pandas as pd
data = pd.read_csv('C:/MSU/3kurs/6sem/statprak/Table.csv')
data['Project'] = data['Project'] * 4 #Делаем вес за проект более весомым (в 4 раза весомее обычного дз)
data['Pandas'] = data['Pandas'] * 2  #Баллы за дз Pandas также делаем выше, т.к. это было важное дз за которое было много вопросов на понимание
data = data.fillna(-1) #Пустым ячейкам присваиваем (-1) - люди которые ничего не делали будут отдельно от всех
results = data.iloc[:, 1:].sum(axis=1)
results = pd.concat([data['Student'], results], axis=1)
results = results.rename(columns={0: 'result'})

results
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(results['result'].values.reshape(-1, 1))
results['Cluster'] = kmeans.labels_
cluster_mapping = {0: 'автомат', 1: 'на зачёт', 2: 'пересдача'}
results['Cluster'] = results['Cluster'].replace(cluster_mapping)
results
```
